```python
import pandas as pd
import bs4
import numpy as np
import requests
import re
```

# the website for web-scrapping is https://www.goodreads.com/list/show/9440.100_Best_Books_of_All_Time_The_World_Library_List

# metrics i am going to scrape the data from that website are 
1-name of the book

2-author of book name


3-rating of the book and

4- how many number of people rated the book.

```python
document = requests.get("https://www.goodreads.com/list/show/9440.100_Best_Books_of_All_Time_The_World_Library_List")
print(document)
```

# importing the web scrapper beautifulsoup

```python
from bs4 import BeautifulSoup
docparse = BeautifulSoup(document.content,'html.parser')
```

# let's create empty list's to store the collected data

```python
book_name = []
book_rating = []
author = []
```

```python
print(docparse)
```

```python
name = docparse.find_all('a',attrs = {'class':'bookTitle'})
book_author = docparse.find_all('a',attrs={'class':'authorName'})
rating = docparse.find_all('span',attrs={'class':'minirating'})
```

# iterating through the derived data and append it to empty lists

```python
for i in name:
    book_name.append(i.text)
print(book_name)
```

```python
for i in book_author:
    author.append(i.text)
print(author)
```

```python
for i in rating:
    book_rating.append(i.text)
print(book_rating)
```

# now create a dataframe from the derived or scrapped data

```python
book_data = pd.DataFrame()
book_data['BookName'] = book_name
book_data['BookAuthor'] = author
book_data['BookRating'] = book_rating
book_data['NumRating'] = book_rating
```

```python
book_data.head()
```

# Data Cleaning

# 
we can see from above that bookname has white spaces "\n" in both ends and we need to remove that to make data clean to be used 

```python
book_data['BookName'] = book_data['BookName'].apply(lambda x:re.findall(r'\n(.*?)\n',x)[0])
```

```python
book_data.head()
```

# now we need to store the rating in "BookRating" column and number of rating in "NumRating" column

```python
book_data['BookRating'] = book_data['BookRating'].apply(lambda x:re.findall(r'(.*?) avg rating*', x)[0])
```

```python
book_data.head()
```

```python
book_data['NumRating'] = book_rating
book_data['NumRating'] = book_data['NumRating'].apply(lambda x: x[18:-7])
```

```python
book_data.head()
```

```python
book_data.to_csv('book_data.csv',index=False)
```
